```python
def remove_space(list_headers,characters) : #characters should include ['-',' ']
    new_headers = list()
    for header in list_headers:
        for char in characters:
            if char in header:
                header = header.replace(char,'_')
            header = header
        new_headers.append(header)
    return new_headers

def remove_sp_char(headers,sp_characters): #special characters include ["''",'/',':'']
    new_headers = list()
    for header in headers:
        for char in sp_characters:
            if char == "-" or char == '/':
                header = header.replace(char,'_')
            if char in header:
                header = header.replace(char,'')
            header = header
        new_headers.append(header)
    return new_headers

headers = remove_space(uni_df.columns,[' ',' - '])
headers = remove_sp_char(headers,[':',"'",'/',',','-'])
```

```python
variables = ['applicants_total', 'admissions_total', 'enrolled_total']
titles = ['Number of Applications', 'Number of Admissions', 'Number of Enrollments']
limits = [(0, 30000), (0, 10000), (0, 5000)]

plt.figure(figsize=(20, 6))

for i, var in enumerate(variables, start=1):
    plt.subplot(1, 3, i)
    sns.histplot(uni_df[var], bins=50, color='skyblue')
    plt.title(f'Histogram of {titles[i-1]}\nMean: {uni_df[var].mean():.1f}, Median: {uni_df[var].median():.1f}')
    plt.xlabel(titles[i-1])
    plt.axis([0, limits[i-1][1], 0, 350])
    plt.xticks(rotation=10)
    plt.grid()

plt.show()
```

#### Q 2.) Do students show a preference for universities with a high admission rate? In simpler terms, are students drawn to institutions where admission is more readily accessible?

```python
# Set the figure size
plt.figure(figsize=(16, 6))

# Scatter plot: Applications vs Admissions
plt.subplot(1, 2, 1)
sns.scatterplot(x='applicants_total', y='admissions_total', hue='control_of_institution', data=uni_df)
plt.title('Applications vs Admissions')
plt.xlabel('Number of Applications')
plt.ylabel('Number of Admissions')
plt.grid()

# Scatter plot: Admissions vs Enrollments
plt.subplot(1, 2, 2)
sns.scatterplot(x='admissions_total', y='enrolled_total', hue='control_of_institution', data=uni_df)
plt.title('Admissions vs Enrollments')
plt.xlabel('Number of Admissions')
plt.ylabel('Number of Enrollments')
plt.grid()

# Show the plots
plt.show()
```

In the first scatterplot, we observe a correlation where universities with a substantial number of applications tend to have a high admission rate, and vice versa, with a few exceptions. Notably, some private universities with a high volume of applications exhibit a low acceptance rate, indicating a selective admission process.

In the second scatterplot, a clear trend emerges, showing that a higher number of admissions corresponds to an increased enrollment rate. To enhance the significance of our analysis, we can delve into the admissions/acceptance rate and enrollment rate. Let's proceed to calculate the admission and enrollment rates for the universities.

```python
# Calculate admission rate as a percentage
uni_df['offer_rate'] = (uni_df['admissions_total'] / uni_df['applicants_total']) * 100

# Calculate enrollment rate as a percentage
uni_df['enrollment_rate'] = (uni_df['enrolled_total'] / uni_df['admissions_total']) * 100
```

```python
plt.figure(figsize=(12, 5))
sns.scatterplot(x='applicants_total', y='enrollment_rate', data=uni_df)
plt.xlabel('Number of Applications')
plt.ylabel('Enrollment Rate (%)')
plt.grid()
plt.show()
```

This figure indicates that universities receiving a lower number of applications tend to have a higher enrollment rate. While there are exceptions, this trend is notably strong. Therefore, it can be concluded that a high number of applications does not necessarily imply that a university is preferred among students.

Next, let's address the inquiry concerning the number of admissions.

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(x='offer_rate', y='enrollment_rate', hue='control_of_institution', data=uni_df)
plt.xlabel('Offer Rate')
plt.ylabel('Enrollment Rate')
plt.grid()
plt.show()
```

We observe that, for high acceptance rates, enrollments exhibit considerable variation. This suggests that acceptance rate may not be a decisive factor influencing students' preferences when choosing a university. However, for low acceptance rates, the enrollment rate tends to be consistently above 40%, especially for private universities within this range. This leads us to the conclusion that the majority of public colleges have an acceptance rate exceeding 50%. Additionally, it is noteworthy that when students are accepted to universities with low acceptance rates, they tend to enroll straightforwardly. Further analysis of this phenomenon will be conducted.

```python
# Filter out universities with non-null offer rates
high_acceptance = uni_df[uni_df.offer_rate.notnull()][['name', 'offer_rate', 'enrollment_rate']].nlargest(10, 'offer_rate')
low_acceptance = uni_df[uni_df.offer_rate.notnull()][['name', 'offer_rate', 'enrollment_rate']].nsmallest(10, 'offer_rate')

plt.figure(figsize=(16, 4))

# Plot for Top 10 Universities with Highest Acceptance Rate
plt.subplot(1, 2, 1)
ind = np.arange(len(high_acceptance))
width = 0.30

plt.bar(ind, high_acceptance['offer_rate'], width, label='Acceptance Rate')
plt.bar(ind + width, high_acceptance['enrollment_rate'], width, label='Enrollment Rate')
plt.title('Top 10 Universities with Highest Acceptance Rate')
plt.ylabel('Rates %')
plt.xticks(ind + width / 2, high_acceptance['name'].values, rotation=90)
plt.legend(loc='best')

# Plot for Top 10 Universities with Lowest Acceptance Rate
plt.subplot(1, 2, 2)
ind = np.arange(len(low_acceptance))
width = 0.30

plt.bar(ind, low_acceptance['offer_rate'], width, label='Acceptance Rate')
plt.bar(ind + width, low_acceptance['enrollment_rate'], width, label='Enrollment Rate')
plt.title('Top 10 Universities with Lowest Acceptance Rate')
plt.ylabel('Rates %')
plt.xticks(ind + width / 2, low_acceptance['name'].values, rotation=90)
plt.legend(loc='best')

plt.show()
```

We observe that there is no distinct pattern in the enrollment rates of universities with high admission rates. This lack of a clear trend indicates that students' preferences are not necessarily influenced by how easily they can gain admission.

Before proceeding with the analysis of costs based on states or regions, the data will be examined to determine whether these patterns persist when combining the costs for in-state and out-of-state students. Additionally, a new column will be added to the dataset, providing information about the region where these universities are situatedâ€”a factor that will be useful for subsequent analysis.

```python
uni_df=uni_df.merge(region_df,on='state')
```

```python
def create_living_cost_df(total_price_column, state_type):
    living_cost_df = uni_df[['name', 'enrollment_rate', total_price_column, 'control_of_institution', 'state', 'region']]
    living_cost_df = living_cost_df.rename(columns={total_price_column: 'price_living'})
    living_cost_df['from'] = state_type
    return living_cost_df

# In-State DataFrame
in_state_df = create_living_cost_df('total_price_for_in_state_students_living_on_campus_2013_14', 'In-State')

# Out-of-State DataFrame
out_state_df = create_living_cost_df('total_price_for_out_of_state_students_living_on_campus_2013_14', 'Out-State')

in_out_state_df = pd.concat([in_state_df, out_state_df], ignore_index=True)
```

```python
plt.figure(figsize=(16,6))

#with standard deviation
plt.subplot(1,2,1)
sns.barplot(x='control_of_institution',y='price_living',data=in_out_state_df,hue='from',ci="sd",palette='rocket');
plt.title('''Average and variation of the 
cost of on-campus living (2013-2014)''')
plt.xlabel('')
plt.ylabel('Cost of on-campus living $')

plt.subplot(1,2,2)
sns.scatterplot(x='price_living',y='enrollment_rate',data=in_out_state_df,hue='from',palette='rocket')
plt.title('Cost of On-Campus Living vs Enrollment Rate')
plt.xlabel('Cost of On-Campus Living $')
plt.ylabel('Enrollment Rate');
```

In summary, the analysis has consistently shown that the average cost offered by public universities is lower than that of private universitiesâ€”a confirmation supported by the left figure. Further insight can be derived by examining all costs side by side. For public universities, the average cost for out-of-state students is notably higher than that for in-state students. Conversely, in the case of private universities, the average costs for in-state and out-of-state students are identical.

#### Q 7.) Do students base their university choices on the location within a state with a high GDP per capita?

```python
# Drop the 'code' column from the gdp_df dataframe in-place
state_gdp_df.drop(columns=['code'], inplace=True)

# Merge the univnew_df dataframe with the modified gdp_df dataframe on the 'state' column
uni_df = uni_df.merge(state_gdp_df, on='state', how='left')  # Using 'left' join for clarity
```

```python
uni_df[['state','gdp_million','population']].tail()
```

Given that we have the GDP of each state but intend to work with GDP per capita, it is necessary to calculate this value.

```python
# Calculate GDP per capita and create a new column in uni_df
uni_df['gdp_capita'] = uni_df['gdp_million'] / uni_df['population'] * 1e6

# Group by 'state' and calculate the mean of 'gdp_capita'
gdp_state_df = uni_df.groupby('state')['gdp_capita'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(16, 16))
ax = sns.barplot(x=gdp_state_df.values, y=gdp_state_df.index,  palette='viridis')
ax.set_xlim((0, 180000))
plt.title('GDP per Capita of American States')
plt.xlabel('GDP per Capita $')
plt.grid(axis='x', alpha=0.75)

# Show the plot
plt.show()
```

The District of Columbia boasts the highest GDP per capita, while Louisiana has the lowest. Notably, there is a significant gap between the highest GDP per capita (from the District of Columbia) and the next in line (Alaska), with the difference starting at around $78,000.

To further explore potential correlations, let's investigate whether there is a relationship between GDP per capita and enrollment rate.

```python
# Define constants
target_gdp = 78000

# Plot the scatterplot
plt.figure(figsize=(16, 7))
ax = sns.scatterplot(x='gdp_capita', y='enrollment_rate', data=uni_df)

# Add a vertical line at the target GDP per Capita
ax.axvline(x=target_gdp, color='red', lw=3, linestyle='--')

# Set plot title and labels
plt.title('GDP per Capita vs Enrollment Rate')
plt.grid()
plt.xlabel('GDP per Capita $')
plt.ylabel('Enrollment Rate %')

# Show the plot
plt.show()
```

A red dashed line has been incorporated to demarcate the commencement of the aforementioned GDP gap.

Notably, despite having the highest GDP per capita, the District of Columbia does not correspondingly exhibit the highest enrollment rates.

For a more nuanced understanding of this relationship, let's zoom in on the image, excluding the data point for the District of Columbia.

```python
# Set the figure size and axis limits
plt.figure(figsize=(16, 7))
plt.axis([30000, 80000, 0, 101])

# Plot the scatterplot
sns.scatterplot(x='gdp_capita', y='enrollment_rate', data=uni_df)

# Set grid, title, and labels
plt.grid()
plt.title('GDP per Capita vs Enrollment Rate')
plt.xlabel('GDP per Capita $')
plt.ylabel('Enrollment Rate %')

# Remove spines for a cleaner appearance
sns.despine()

# Show the plot
plt.show()
```

The figure indicates a lack of correlation between a high GDP per capita and a high enrollment rate. Consequently, we can conclude that students do not exhibit a preference for a university based on the GDP per capita of the state in which the university is located.
